```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://books.toscrape.com/"
```

```python
soup = BeautifulSoup(requests.get(url).content, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
books_list = []
for book in soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
  img='http://books.toscrape.com/'+book.find("img", class_="thumbnail")["src"]
  title=book.find("h3").text
  price=book.find("p", class_="price_color").text
  rating=book.find("p")["class"][1]
  books_list.append([title,rating,price,img])
```

```python
for i in books_list:
  print(i)
```

```python
df=pd.DataFrame(books_list,columns=["Title","Rating","Price","Image"])
df
```

```python
# save csv file
df.to_csv("books.csv")
```
