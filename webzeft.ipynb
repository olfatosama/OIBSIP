```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.w3schools.com/"
```

```python
soup = BeautifulSoup(requests.get(url).content, "html.parser")
soup
```

```python
soup.prettify()
```

```python
links = soup.find_all("a")
links
```

```python
lst=[]
lst1=[]
for link in links:
        href = link.get('href')
        title = link.get('title')  # Extract the title attribute directly

        # Check if href is not None and starts with a valid URL or path
        if href and (href.startswith('http') or href.startswith('/')):
            # If it's a relative URL, convert it to an absolute URL
            if href.startswith('/'):
                href = 'https://www.w3schools.com' + href

            # Only print if a title is present
            if title:
                print(f"Title: {title} \nURL: {href}\n")
                # lst.append(f"Title: {title} \nURL: {href}\n")
                lst1.append({'title':title,'url':href})
else:
    print('Failed to retrieve the webpage.')
```
